```python
import base64
import httpx
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

image_url_1 = "https://ts1.tc.mm.bing.net/th/id/R-C.f18fc940ed5001e9a4a0d10b4b8e5290?rik=QXRPDMI2J%2fZaKA&riu=http%3a%2f%2fpic.baike.soso.com%2fp%2f20130613%2f20130613113407-1702189926.jpg&ehk=VW0ihZazqrqTdBrufI6kjT19d1HOqe6h4hKoVMTobSA%3d&risl=&pid=ImgRaw&r=0"
image_url_2 = "https://img.shetu66.com/2023/04/25/1682391068745168.png"

# OpenAIbase64
def url_to_base64(url):
    response = httpx.get(url)
    return f"data:image/jpeg;base64,{base64.b64encode(response.content).decode('utf-8')}"

image_data_1 = url_to_base64(image_url_1)
image_data_2 = url_to_base64(image_url_2)

model = ChatOpenAI(
    model="llama3.2-vision",  
    base_url="http://localhost:11434/v1/",
    api_key="sk-no-key-required",
    temperature=0.7
)


# 
message = HumanMessage(
    content=[
        {"type": "text", "text": ""},
        {"type": "image_url", "image_url": {"url": image_data_1}},  # base64
        {"type": "image_url", "image_url": {"url": image_data_2}}
    ]
)

# 
try:
    response = model.invoke([message])
    print(response.content)
except Exception as e:
    print(f"API: {str(e)}")
```

## YAML 

```python
from langchain.output_parsers import YamlOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field


# Define your desired data structure.
class Joke(BaseModel):
    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")


model = ChatOpenAI(
    model="llama3.1:latest",  
    base_url="http://localhost:11434/v1/",
    api_key="sk-no-key-required",
    temperature=0.7
)

# And a query intented to prompt a language model to populate the data structure.
joke_query = "Tell me a joke."

# Set up a parser + inject instructions into the prompt template.
parser = YamlOutputParser(pydantic_object=Joke)

prompt = PromptTemplate(
    template="Answer the user query.\n{format_instructions}\n{query}\n",
    input_variables=["query"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

chain = prompt | model | parser

chain.invoke({"query": joke_query})
```
